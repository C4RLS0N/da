```python
#111111111111111111111111
#Библиотеки
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt

#22222222222222222222
# Загрузка датасета
dataset, info = tfds.load('oxford_iiit_pet:3.*.*', with_info=True)

#3333333333333
# Функция для преобразования изображения
def normalize(input_image, input_mask):
    input_image = tf.cast(input_image, tf.float32) / 255.0
    input_mask -= 1
    return input_image, input_mask

# Функция для изменения размера изображения
def resize(input_image, input_mask):
    input_image = tf.image.resize(input_image, (128, 128))
    input_mask = tf.image.resize(input_mask, (128, 128))
    return input_image, input_mask

# Преобразование датасета
data = dataset['train'].map(lambda x: (x['image'], x['segmentation_mask']))
data = data.map(resize)
data = data.map(normalize)

#444444444444
# Функции для создания слоев свертки и пулинга
def conv2d_block(input_tensor, n_filters, kernel_size=3, batchnorm=True):
    # первый сверточный слой
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size),
                               kernel_initializer='he_normal', padding='same')(input_tensor)
    if batchnorm:
        x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)
    # второй сверточный слой
    x = tf.keras.layers.Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size),
                               kernel_initializer='he_normal', padding='same')(x)
    if batchnorm:
        x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation('relu')(x)
    return x

def encoder_block(inputs, n_filters=64, pool_size=(2,2), dropout=0.3):
    f = conv2d_block(inputs, n_filters)
    p = tf.keras.layers.MaxPooling2D(pool_size=(2,2))(f)
    p = tf.keras.layers.Dropout(dropout)(p)
    return f, p

def decoder_block(inputs, conv_output, n_filters=64, kernel_size=3, dropout=0.3):
    up = tf.keras.layers.Conv2DTranspose(n_filters, kernel_size, strides=(2,2), padding='same')(inputs)
    merge = tf.keras.layers.concatenate([conv_output, up], axis=3)
    c = conv2d_block(merge, n_filters, kernel_size=kernel_size, batchnorm=False)
    c = tf.keras.layers.Dropout(dropout)(c)
    c = conv2d_block(c, n_filters, kernel_size=kernel_size, batchnorm=False)
    return c

# Функция для создания модели Unet
def create_unet():
    # Входной слой
    inputs = tf.keras.layers.Input(shape=(128, 128, 3,))

    # Encoder
    f1, p1 = encoder_block(inputs, n_filters=64, pool_size=(2,2), dropout=0.3)
    f2, p2 = encoder_block(p1, n_filters=128, pool_size=(2,2), dropout=0.3)
    f3, p3 = encoder_block(p2, n_filters=256, pool_size=(2,2), dropout=0.3)
    f4, p4 = encoder_block(p3,n_filters=512, pool_size=(2,2), dropout=0.3)

# Bottleneck
    bn = conv2d_block(p4, n_filters=1024)

# Decoder
    c1 = decoder_block(bn, conv_output=f4, n_filters=512, kernel_size=3, dropout=0.3)
    c2 = decoder_block(c1, conv_output=f3, n_filters=256, kernel_size=3, dropout=0.3)
    c3 = decoder_block(c2, conv_output=f2, n_filters=128, kernel_size=3, dropout=0.3)
    c4 = decoder_block(c3, conv_output=f1, n_filters=64, kernel_size=3, dropout=0.3)

# Выходной слой
    outputs = tf.keras.layers.Conv2D(3, (1, 1), activation='softmax')(c4)

# Создание модели
    model = tf.keras.models.Model(inputs=[inputs], outputs=[outputs])

    return model


#Создание модели
model = create_unet()

#Компиляция модели
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

#Обучение модели
history = model.fit(data.batch(32), epochs=20)

#5555555555555555
# Визуализация входного изображения, истинной маски и предсказанной маски для первых 5 изображений из датасета
images, masks = next(iter(data.batch(5)))

predicted_masks = model.predict(images)

plt.figure(figsize=(15,15))
for i in range(5):
    plt.subplot(5,3,i*3+1)
    plt.imshow(images[i])
    plt.title("Input Image")
    plt.axis('off')

    plt.subplot(5,3,i*3+2)
    plt.imshow(masks[i])
    plt.title("True Mask")

    plt.subplot(5,3,i*3+3)
    plt.imshow(predicted_masks[i])
    plt.title("Predicted Mask")
    plt.axis('off')

plt.show()

#666666666666666
#Визуализация структуры модели в виде графа
tf.keras.utils.plot_model(model, show_shapes=True, dpi=64)
```
